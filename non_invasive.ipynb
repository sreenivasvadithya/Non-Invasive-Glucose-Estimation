```python
import requests
import pandas as pd

firebase_url = "https://glucosemonitor-77d2a-default-rtdb.asia-southeast1.firebasedatabase.app/readings.json"

response = requests.get(firebase_url)

if response.status_code == 200:
    data = response.json()
    print("Full Firebase Response:")
    print(data)


    if isinstance(data, dict):
        # Get the latest entry (by Firebase push key order)
        latest_key = list(data.keys())[-1]
        latest_data = data[latest_key]

        ir_voltage = latest_data.get("IR_Voltage")
        status = latest_data.get("Status", "Unknown")

        if ir_voltage is not None:
            print("Live IR Voltage:", ir_voltage)
            print("Status:", status)

            # Predict glucose
            input_df = pd.DataFrame([[ir_voltage]], columns=['NIR_Voltage'])
            predicted = model.predict(input_df)
            print(f"Predicted Glucose: {predicted[0]:.2f} mg/dL")
        else:
            print("IR_Voltage not found in latest data.")
    else:
        print("Unexpected Firebase data format.")
else:
    print("Failed to fetch from Firebase:", response.status_code)
```

```python
import time

while True:
    response = requests.get(firebase_url)
    if response.status_code == 200:
        data = response.json()
        latest_key = list(data.keys())[-1]
        latest_data = data[latest_key]
        ir_voltage = latest_data.get("IR_Voltage")

        if ir_voltage:
            input_df = pd.DataFrame([[ir_voltage]], columns=['NIR_Voltage'])
            predicted = model.predict(input_df)
            print(f"IR: {ir_voltage:.4f} → Glucose: {predicted[0]:.2f} mg/dL\n")
    time.sleep(5)
```
